_Loto fácil is a Brazilian lottery where the winner hits 15 out of 25 possible numbers._

- _Create DataFrame_

- _Multi-Index_
- _Sort Data_

```python
# import library
import pandas as pd
import glob
```

```python
# reading one of the files
pd.read_excel("files/loto_facil_2020_sorteio.xlsx").head(9)
```

```python
# Files consolidation
all_files = glob.glob("files/loto_facil_*.xlsx")

li = []

for filename in all_files:
    df = pd.read_excel(filename, index_col=None, header=0)
    header = df.iloc[5] # get names to columns
    df = df[6:] # get data of 6th index to down
    df = df.rename(columns = header) # Rename columns
    df = df.sort_index(ascending=False) # Set index to descending
    df["Data"] = pd.to_datetime(df["Data"], format="%d/%m/%Y") # Date format change
    li.append(df)
    
drawn_df = pd.concat(li, axis=0, ignore_index=True)
```

```python
# read columns
drawn_df.columns
```

```python
# Rename columns
drawn_df.columns = ['loto_facil', 'date', 'ball_1', 'ball_2', 'ball_3', 'ball_4', 'ball_5',
                    'ball_6', 'ball_7', 'ball_8', 'ball_9', 'ball_10', 'ball_11', 'ball_12', 
                    'ball_13', 'ball_14', 'ball_15']
#Set multi index
drawn_df.set_index(['loto_facil', 'date'], inplace = True)
```

```python
drawn_df = drawn_df.sort_index(ascending=True)
```

```python
# read first and last line
drawn_df.loc[[1, len(drawn_df)]]
```

```python
# read the last five line
drawn_df.tail()
```

_**Note:** Describing the data frame, we can see the most drawn number in each position._

```python
# Describe dataframe
drawn_df.describe()
```

_Transforming the data frame to array._

```python
matrix = drawn_df.values
matrix
```

_bubble_sort function to array sorting_

```python
def bubble_sort(alist):
    for passnum in range(len(alist)-1, 0, -1):
        for i in range(passnum):
            if alist[i] > alist[i+1]:
                temp = alist[i]
                alist[i] = alist[i+1]
                alist[i+1] = temp


i = 0
while i < len(matrix):
    bubble_sort(matrix[i])
    i = i + 1
```

_Ordered array._

```python
matrix
```

_**Note:** Creating the `lottery_df` using `matrix` and `drawn_df`'s index._

```python
# Create dataframe
lottery_df = pd.DataFrame(matrix, index=drawn_df.index, 
                           columns=list(['num_1', 'num_2', 'num_3', 'num_4','num_5',
                                         'num_6', 'num_7','num_8', 'num_9', 'num_10',
                                         'num_11', 'num_12', 'num_13','num_14', 'num_15']))
```

_Transforming array into two columns_

```python
def array_into_2columns(alist):
    li = []
    for ind in range(len(alist)):
        for col in range(len(alist[0])):
            li.append([ind+1,alist[ind][col]])
            
    return li

two_columns = array_into_2columns(matrix)
```

```python
two_columns_df = pd.DataFrame(two_columns, columns=['loto_facil','number'])
two_columns_df.set_index('loto_facil', inplace=True)
two_columns_df.tail()
```

_Count numbers._

```python
most_drawn = two_columns_df['number'].value_counts().to_frame(name="Count numbers")
most_drawn
```
